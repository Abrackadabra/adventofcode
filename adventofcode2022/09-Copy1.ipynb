```python
def touching(a, b, dirs):
  for dir in dirs:
    if a + dir == b:
      return True
  return False


def how_to_move(a, b):
  if touching(a, b, DIRS8) or a == b:
    return ORIGIN
  
  for maybe_dir in DIRS4:
    if touching(a, b + maybe_dir, DIRS4):
      return maybe_dir
    
  for maybe_dir in DIRS_DIAG:
    if touching(a, b + maybe_dir, DIRS4):
      return maybe_dir
    
  for maybe_dir in DIRS_DIAG:
    if touching(a, b + maybe_dir, DIRS_DIAG):
      return maybe_dir
    
  raise Exception()


parts = [ORIGIN for i in range(10)]

diffbs = set()
diffbs.add(parts[-1])

for line in lines:
  if m := re.findall(r'^(.) (\d+)$', line):
    direc = m[0][0]
    steps = int(m[0][1])
    
    head_move = POINTS_UP[direc]
      
    for i in range(steps):
      parts[0] += head_move
      
      for j in range(9):
        parts[j + 1] += how_to_move(parts[j], parts[j + 1])
              
      diffbs.add(parts[-1])
    
  else:
    raise Exception()
    
len(diffbs)
```
