```python
def intersect(cube1, cube2):
  x1 = max(cube1[0], cube2[0])
  x2 = min(cube1[1], cube2[1])
  y1 = max(cube1[2], cube2[2])
  y2 = min(cube1[3], cube2[3])
  z1 = max(cube1[4], cube2[4])
  z2 = min(cube1[5], cube2[5])
  
  if x1 < x2 and y1 < y2 and z1 < z2:
    return x1, x2, y1, y2, z1, z2
  
def within(cube1, cube2):
  return all([
    cube1[0] >= cube2[0] and cube1[1] <= cube2[1],
    cube1[2] >= cube2[2] and cube1[3] <= cube2[3],
    cube1[4] >= cube2[4] and cube1[5] <= cube2[5],
  ])
```

```python
def split(cube1, cube2):
  xs = sorted(unique([cube1[0], cube1[1], cube2[0], cube2[1]]))
  ys = sorted(unique([cube1[2], cube1[3], cube2[2], cube2[3]]))
  zs = sorted(unique([cube1[4], cube1[5], cube2[4], cube2[5]]))
  cubes1 = []
  cubes2 = []
  cubes3 = []
  for x1, x2 in zip(xs[:-1], xs[1:]):
    for y1, y2 in zip(ys[:-1], ys[1:]):
      for z1, z2 in zip(zs[:-1], zs[1:]):
        cube = (x1, x2, y1, y2, z1, z2)
        in1 = within(cube, cube1)
        in2 = within(cube, cube2)
        if in1 and not in2:
          cubes1.append(cube)
        if in1 and in2:
          cubes2.append(cube)
        if not in1 and in2:
          cubes3.append(cube)
  return cubes1, cubes2, cubes3

def volume(cube):
  return (cube[1] - cube[0]) * (cube[3] - cube[2]) * (cube[5] - cube[4])

def bb(cube1, cube2):
  return (
    min(cube1[0], cube2[0]),
    max(cube1[1], cube2[1]),
    min(cube1[2], cube2[2]),
    max(cube1[3], cube2[3]),
    min(cube1[4], cube2[4]),
    max(cube1[5], cube2[5]),
  )

def can_join(cube1, cube2):
  c = bb(cube1, cube2)
  cubes1, cubes2, cubes3 = split(cube1, cube2)
#   print(cubes1, cubes2, cubes3)
  if len(cubes2) == 0 and sum(amap(volume, cubes1)) + sum(amap(volume, cubes3)) == volume(c):
    return c
```

```python
def merge(cubes):
  cubes = sorted(cubes)
  while True:
    x = 0
    i = 0
    while i < len(cubes) - 1:
      cube1, cube2 = cubes[i:i+2]
      if c := can_join(cube1, cube2):
        cubes.pop(i + 1)
        cubes[i] = c
        x += 1
      else:
        i += 1
    if x == 0:
      break
  return cubes
```

```python
@dataclasses.dataclass
class Node:
  bounds: typing.Tuple[int, int, int, int, int, int]
    
  children_bounds: typing.List[typing.Tuple[int, int, int, int, int, int]]
  children: typing.List['Node']

  full: typing.Optional[bool]
  
  def __init__(self, bounds, fill):
    self.bounds = bounds
    self.full = fill
    self.children_bounds = []
    self.children = []
    
  def spawn_children(self, fill):
    if volume(self.bounds) <= 1:
      return
    
    xl, xr, yl, yr, zl, zr = self.bounds
    xm = (xl + xr) // 2
    ym = (yl + yr) // 2
    zm = (zl + zr) // 2
    
    for child in [
      (xl, xm, yl, ym, zl, zm),
      (xl, xm, yl, ym, zm, zr),
      (xl, xm, ym, yr, zl, zm),
      (xl, xm, ym, yr, zm, zr),
      (xm, xr, yl, ym, zl, zm),
      (xm, xr, yl, ym, zm, zr),
      (xm, xr, ym, yr, zl, zm),
      (xm, xr, ym, yr, zm, zr),
    ]:
      if volume(child) > 0:
        self.children_bounds.append(child)
        self.children.append(Node(child, fill))
        
  def count(self):
    res = 1
    if self.full is True:
      return res
    elif self.full is False:
      return res
    elif self.full is None:
      pass
    
    for child in self.children:
      res += child.count()
    
    return res
    
  
  def volume(self):
    if self.full is True:
      return volume(self.bounds)
    elif self.full is False:
      return 0
    elif self.full is None:
      pass
    
    res = 0
    
    for child in self.children:
      res += child.volume()
    
    return res
  
  def add(self, cube):
    if cube == self.bounds:
      self.full = True
      self.children_bounds.clear()
      self.children.clear()
      return

    if self.full is True:
      return
    elif self.full is False:
      self.full = None
      self.spawn_children(False)
    elif self.full is None:
      pass
    
    for i in range(len(self.children_bounds)):
      if (intersection := intersect(cube, self.children_bounds[i])) is None:
        continue
            
      self.children[i].add(intersection)

    if all([child.full is True for child in self.children]):
      self.full = True
      self.children_bounds.clear()
      self.children.clear()
  
  def remove(self, cube):
    if cube == self.bounds:
      self.full = False
      self.children_bounds.clear()
      self.children.clear()
      return

    if self.full is True:
      self.full = None
      self.spawn_children(True)
    elif self.full is False:
      return
    elif self.full is None:
      pass
    
    for i in range(len(self.children_bounds)):
      if (intersection := intersect(cube, self.children_bounds[i])) is None:
        continue
      
      self.children[i].remove(intersection)
      
    if all([child.full is False for child in self.children]):
      self.full = False
      self.children_bounds.clear()
      self.children.clear()
```
