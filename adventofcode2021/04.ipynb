```python
lines, groups = linesplit(raw_input, verbose=True)
```

```python
nums = amap(int, groups[0][0].split(','))

aa = []

for i in groups[1:]:
    a = [amap(int, j.strip().split()) for j in i]
    a = array(a)
    aa.append(a)
```

```python
marked = set()
boards_won = set()
for i in nums:
    marked.add(i)
    for a_i in range(len(aa)):
        if a_i in boards_won:
            continue
        a = aa[a_i]
        found = False
        
        for j in range(a.shape[0]):
            if all([jj in marked for jj in a[j, :]]):
                found = True
        for j in range(a.shape[1]):
            if all([jj in marked for jj in a[:, j]]):
                found = True
        
        if found:
            s = 0
            for j in a:
                for jj in j:
                    if jj not in marked:
                        s += jj
            print(i, s)
            s *= i
            print(s)
            boards_won.add(a_i)
```
