```python
resi = 0
resa = []

d = {}

for line in lines:
    for k, v in re.findall(r'(\S+):(\S+)', line):
        d[k] = v
    
    if line == '':
#         print(set(list(d.keys())))
        ok = True
        if set(list(d.keys())) & fields != fields:
            ok = False
        
        if ok:
            s = d['byr']
            ok &= (
                re.findall(r'\d{4}', s)[0] == s
                and (1920 <= int(s) <= 2002)
            )
            print(s, ok)
        
            s = d['iyr']
            ok &= (
                (2010 <= int(s) <= 2020)
                and re.findall(r'\d{4}', s)[0] == s
            )
            print(s, ok)

            s = d['eyr']
            ok &= (
                (2020 <= int(s) <= 2030)
                and re.findall(r'\d{4}', s)[0] == s
            )
            print(s, ok)

            s = d['hgt']
            ok &= (
                (re.findall(r'\d{3,}cm', s) == [s] and (150 <= int(s[:-2]) <= 193))
                or (re.findall(r'\d{2,}in', s) == [s] and (59 <= int(s[:-2]) <= 76))
            )
            print(s, ok)

            s = d['hcl']
            ok &= (
                (re.findall(r'#[0-9a-f]{6}', s) == [s])
                or (re.findall(r'#[a-f]{6}', s) == [s])
            )
            print(s, ok)

            s = d['ecl']
            ok &= (
                (re.findall(r'amb|blu|brn|gry|grn|hzl|oth', s) == [s])
            )
            print(s, ok)

            s = d['pid']
            ok &= (
                (re.findall(r'[0-9]{9}', s) == [s])
            )
            print(s, ok)

        if ok:
            resi += 1
        
        d = {}
        
resi
```
