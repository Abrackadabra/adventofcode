```python
line = lines[0]

def ev(line):

    s = str(line)

    while True:
        ms = re.findall(r'(?:\(|^)[^\(\)]+(?:\)|$)', s)
        
        if s.isdigit():
            break
        
        for i in ms:
            ds = amap(int, re.findall(r'\d+', i))

            x = ds[0]

            for op, d in zip(re.findall(r'[\+\*]', i), ds[1:]):
                if op == '+':
                    x += d
                elif op == '*':
                    x *= d
                else:
                    raise Exception()

            s = s.replace(i, str(x))

            print(i, x, s)
        
    return int(s)
    
ev(lines[0])
```
