```python
def get_codes(s):
    res = {}
    for l, r in itertools.combinations(range(len(s) + 1), 2):
        x = 0
        for i in s[l:r]:
            x *= 3
            if i == 'a':
                x += 1
            else:
                x += 2
        res[(l, r)] = x
    return res

# codes(matches[0])
```

```python
# dp = zeros(100, dtype=int)
# ddp = zeros_like(dp)


def recmatch(codes, s, gmem, l, r, v):
#     global dp, ddp
#     dp = zeros(len(s) + 1, dtype=int)
#     ddp = zeros(len(ss) + 1,)
    
    if (codes[l, r], v) in gmem:
        return gmem[(codes[l, r], v)]
    
    for option in resd[v]:
        if len(option) == 1:
            t, vv = option[0]

            if t == 0 and s[l:r] == vv:
                gmem[(codes[l, r], v)] = True
                return True
            if t == 1 and recmatch(codes, s, gmem, l, r, vv):
                gmem[(codes[l, r], v)] = True
                return True
            
            continue

        assert set([i[0] for i in option]) == {1}

        dp = zeros(len(s) + 1)
        ddp = zeros_like(dp)
        dp[l] = 1

        for _, vv in option:
            ddp *= 0

            for end2 in range(l + 1, r + 1):
                for st2 in where(dp[l:end2] == 1)[0]:
                    if recmatch(codes, s, gmem, st2 + l, end2, vv):
                        ddp[end2] = 1
            
            dp, ddp = ddp, dp

        if dp[r] == 1:
            gmem[(codes[l, r], v)] = True
            return True
    
    gmem[(codes[l, r], v)] = False
    return False
```
